```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/analytics-club-iitm/Summer-School-2022/main/Task_1/data/data.csv')
```

```python
df
```

```python
df.drop_duplicates(inplace=True,keep='first')
```

```python
Grades=df.iloc[:,5].unique()
```

```python
#similar thing can be done for Zone, Education,Remarks etc.
y = []
for i in Grades:
  y.append(df.Grade.value_counts()[i])
plt.bar(Grades, y)
```

```python
plt.title("AGE OF EMPLOYEES")
plt.hist(df.Age,bins=np.arange(20,55,5))
```

```python
# similar thing can be grouped by Education(instead of Grade) as well
plt.figure(figsize=(10,15));
plt.title('SALARY vs TENURE')

#DATASET PROBLEM:Tenure are not exactly float values ( 1.<space>6, 1<space>.6 etc found).
sns.scatterplot(x=df.iloc[:,14],y=df.Tenure,hue=df.Grade);
```
